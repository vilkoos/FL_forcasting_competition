# 0 - setup

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from plotnine import *  # plotnine is a Python ggplot2 clone
```

# 1 - The data

```python
dat = pd.read_csv("https://raw.githubusercontent.com/levvers/dsemr/master/data/week2/sealevel.csv")
print(dat.shape)
dat.head()
```

```python
dat.tail(2)
```

```python
dat.info()
```

dat contains 240 rows, each row represents the mean sea-level during a month.

Each row has 3 measurements
- the **year** (ranging from 1995 to 2014, all 20 years present)
- the **month** within the year (1= January etc, all months present for all years)
- the **sea level** in mm(?) relative to the mean of January 1995

There are no missing or unexpected values in the data.

For convenience lets replace the year/month pair with one measurement-number nr  with nr = (year - 1995)*12 + month

```python
dat['nr'] = (dat['year'] - 1995)*12 + dat['month']

dat.head()
```

# 1.1 - data exploration

Lets visualize the data to see if there is a pattern

```python
ggplot(data=dat, mapping=aes(x='nr', y='sealevel')) + \
    geom_line(color='blue')
```

Overall sea levels are rising about 70 mm over the 20 years.  
There is a clear seasonal component in the figure. We can count 20 peaks, so we may assume a yearly cycle.

Lets fit a smooth line to the figure to check if there is a linear trend present

```python
ggplot(mapping=aes(x='nr', y='sealevel'), data=dat) + \
    geom_point(color='blue', alpha=0.2) + \
    geom_smooth(color='red')
```

The line is almost, but not perfectly straight, lets add a linear model to our figure to see how large the deviation is.

```python
ggplot(mapping=aes(x='nr', y='sealevel'), data=dat) + \
    geom_smooth(method='lm',color='blue',fill='yellow') + \
    geom_smooth(color='red')
```

The red line does not depart far form the strictly linear blue line, so the red smooth trend line can be considered as linear also (furthermore the blue line has a yellow "confidence interval" around it, the red line variation is within the yellow zone).

## 1.2 - the linear component

Lets get the intercept and the slope of the blue line.

```python
# this is the standard import if you're using "formula notation" (similar to R)
import statsmodels.formula.api as smf

# create a fitted model in one line
lm = smf.ols(formula='sealevel ~ nr', data=dat).fit()

# print the coefficients
(my_intercept, my_slope) = lm.params
print(my_intercept)
print(my_slope)
```

Lets add the linear prediction, and the rest that is not explained by the prediction to the data.

```python
def get_linear_component(xnr):
    return my_intercept + (xnr * my_slope)

dat['linear'] = dat['nr'].apply(get_linear_component)
dat['nonLinRest'] = dat['sealevel'] - dat['linear']

dat.head()
```

Visualize to check if there is a pattern in the non-linear rest.

```python
ggplot(data=dat, mapping=aes(x='nr',y='nonLinRest')) + \
    geom_line(color='blue') + \
    geom_smooth(color='red')
```

The blue line shows a yearly cycle with a deviation of about +/- 10 mm.  
The red line hints at a regular 12 to 14 year cycle with a deviation of about +/- 2 m.  

At month nr=80  the red line sealevel is  1 mm.   
At month nr=160 the red line sealevel is -2 mm.   

So the periodicity may be estimated as 80 month.
The slope of the red line is -3mm/80 months = -0.0375.

## 1.3 the mysterious red line effect

Lets assume that the red line oscillates every 80 months and has a slope of +/- 0.0375.

We now can filter out the systematic effects of the red line.

```python
def get_red_effect(nr):
    #slope = 0.0355
    slope = 0.0375
    if nr < 80: return -2 + nr*slope
    elif nr < 160: return 1 - (nr-80)*slope
    elif nr < 240: return -2 + (nr-160)*slope
    else: return 1 - (nr-240)*slope
    #NOTE values of nr > 320 will not occur 

dat['red_effect']  = dat['nr'].apply(get_red_effect)
dat['red_sea_rest']= dat['nonLinRest'] - dat['red_effect']

dat.head()
```

Lets see if correction for the red line effect on the unexplained sealevel variation improved matters

```python
ggplot(data=dat, mapping=aes(x='nr',y='red_sea_rest')) + \
    geom_line(color='blue') + \
    geom_smooth(color='red')
```

## 1.4 - the seasonal effect

For the 20 year lets have a look at the mean of each month (look at the min and max monthly values also)

```python
from pandasql import *

levels_per_month = sqldf("""
SELECT month,
    min(red_sea_rest) as min,
    avg(red_sea_rest) as mean,
    max(red_sea_rest) as max
FROM dat
GROUP BY month
""")

levels_per_month
```

Let visualize the results

```python
ggplot(data=levels_per_month , mapping=aes(x='month')) + \
       geom_line(mapping=aes(y='min')) + \
       geom_line(mapping=aes(y='mean'), color='blue') + \
       geom_line(mapping=aes(y='max')) + \
       geom_point(mapping=aes(y='mean'),color='red')
```

The mean line is pleasantly smooth (there is only a small irregularity in month 4).  
Note that the figure rises 4 months, and declines 8 month (it is symmetrical, it is not a sine).

We will correct for seasonality by adding the months mean to the already explained  sea level variation. 

```python
def get_seasonal_effect(month):
    if   month ==  1: return -0.05
    elif month ==  2: return -1.13
    elif month ==  3: return -2.12
    elif month ==  4: return -2.46
    elif month ==  5: return -4.26
    elif month ==  6: return -4.59
    elif month ==  7: return -2.68
    elif month ==  8: return  0.57
    elif month ==  9: return  4.94
    elif month == 10: return  7.82
    elif month == 11: return  6.54
    elif month == 12: return  3.32
    else: return 999999

dat['seasonal'] = dat['month'].apply(get_seasonal_effect)

dat['predict'] = dat['linear'] + dat['red_effect'] + dat['seasonal']
dat['residu'] = dat['sealevel'] - dat['predict']

dat.head()
```

let have a look at the residuals to see if prediction is about correct.

```python
ggplot(data=dat, mapping=aes(x='nr', y='residu')) + \
    geom_point(color='blue')
```

```python
nr_of_outliers = sqldf(""" SELECT COUNT(*) FROM dat WHERE residu < -5 OR residu > 5 """)

nr_of_outliers
```

So say 85% of predictions are wrong by 5mm or less.


# 2 - predict levels for January 2015 to June 2018

Using the above we can construct a prediction function.

```python
def predict_sealevel(nr,month):
    linear     = get_linear_component(nr)
    red_effect = get_red_effect(nr)
    seasonal   = get_seasonal_effect(month)
    return linear + red_effect + seasonal
```

Make a table with the nr, year and the month we want to predict

```python
nr = 240
pred = []
for year in range(2015,2019):
    for month in range(1,13):
        nr = nr + 1
        pred.append([nr,year,month])
        if year == 2018 and month == 6: break

            
column_names = ['nr','year','month']
predictions = pd.DataFrame(pred, columns=column_names)
predictions.head()
```

```python
predictions.tail()
```

We can now a construct a which we can apply to the predictions dataframe to add a column with predicted sea level values.

```python
def get_prediction(nr):
    month = nr % 12
    if month == 0: month = 12
    return round(predict_sealevel(nr,month),1)

predictions['prediction'] = predictions['nr'].apply(get_prediction)

predictions
```

Lets visualize these results

```python
ggplot(data=predictions, mapping=aes(x='nr', y='prediction')) + \
    geom_line(color='blue')
```

# 4 - save the results to a .xlsx file

Use the [excel writing capabilities](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.to_excel.html) of pandas

```python
writer = pd.ExcelWriter('sealevels.xlsx')
predictions.to_excel(writer,columns=['year','month','prediction'],index=False)
writer.save()
```

Humm . . . it looks like my results are lagging reality by one month.

On hindsight, it would have been better to use the last know actual value as a starting point for my prediction.  
Lets implement the last idea.

```python
def predict_sealevel(nr,month):
    # use the fact that the last know measurement was nr=240 => sealevel=74.0
    last_known = 74.0
    linear     = get_linear_component(nr-240) # calculate growth after month 240
    red_effect = get_red_effect(nr)
    seasonal   = get_seasonal_effect(month)
    return last_known + linear + red_effect + seasonal

def get_new_prediction(nr):
    month = nr % 12
    if month == 0: month = 12
    return round(predict_sealevel(nr,month),1)

predictions['new_pred'] = predictions['nr'].apply(get_new_prediction)

predictions
```

```python
ggplot(data=predictions, mapping=aes(x='nr')) + \
    geom_line(mapping=aes(y='prediction'), color='blue') + \
    geom_line(mapping=aes(y='new_pred'), color='red') 

```

```python
writer = pd.ExcelWriter('new_sealevels.xlsx')
predictions.to_excel(writer,columns=['year','month','new_pred'],index=False)
writer.save()
```
